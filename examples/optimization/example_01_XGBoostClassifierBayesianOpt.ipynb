```python
help(XGBoostClassifierBayesianOpt)
```

```python
# loading data
df = pd.read_csv("data/clf_data.csv")
df.head()
```

```python
# define X, y
y = df.CLASS.values
X = df.drop(["CLASS"], axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.2,
                                                    shuffle=True,
                                                    stratify=y,
                                                    random_state=1367)
```

```python
# define bayesian optimizer 
xbo = XGBoostClassifierBayesianOpt(n_iter=10,
                                   init_points=5,
                                   metrics="auc")
```

```python
# train the optimizer on train set
xbo.fit(X_train, y_train)
```

```python
# optimization results (or xbo.optimization_results_)
xbo.get_optimization_results()
```

```python
# best performance (or xbo.best_performance_)
xbo.get_best_performance()
```

```python
# tuned params (or xbo.best_params_)
xbo.get_best_params()
```

```python
# optimizer object (or xbo.optimizer_)
xbo.get_optimizer()
```

```python
# optimization params boundaries or (xbo.pbounds)
xbo.get_pbounds()
```
